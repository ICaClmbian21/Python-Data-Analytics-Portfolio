```python
import pandas as pd
import urllib
import os
import zipfile
import json
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# 1. Data Acquisition

```python
# Navigate to EIA page containing bulk manifest txt file, which contains a JSON object.
# Convert JSON object into a pandas DataFrame.

path = urllib.urlopen('http://api.eia.gov/bulk/manifest.txt').read()
manifest_data = pd.read_json(path)
manifest_data
```

```python
# Examine dict contents for one sample entry in manifest DataFrame.

manifest_data.ix['TOTAL']['dataset']
```

```python
# Create dict to store hyperlinks to each energy-related category's zip file by pulling URL from manifest DataFrame.

zip_paths = {}
for x in manifest_data['dataset']:
    zip_paths[x['data_set']] = x['accessURL']
```

```python
# Verify that all URLS were stored properly.

for dataset_name in zip_paths:
    print dataset_name,
    print zip_paths[dataset_name]
```

```python
# Create a simple list to store names of each dataset.

dataset_names = []
for dataset_name in zip_paths:
    dataset_names.append(dataset_name)
print dataset_names
```

```python
# Create variable containing local path to folder where zip files will be stored.

local_path = '/Users/adamweber/Google Drive/Data Analytics Files/EIA Zips'
```

```python
# Retrieve each dataset's zip file from URL and store on local disk.

for dataset_name in dataset_names:
    filename = dataset_name + '.zip'
    filename_path = os.path.join(local_path, filename)
    urllib.urlretrieve(zip_paths[dataset_name], filename_path)
```

```python
# Perform extraction from each zip file, which consists of a single .txt file.
# These files contain dataset, broken into numerous, nested JSON objects.

os.chdir(local_path)
for dataset_name in dataset_names:
    zip_name = dataset_name + '.zip'
    txt_name = dataset_name + '.txt'
    f = zipfile.ZipFile(zip_name)
    f.extract(txt_name)
```

# 2. Data Loading/Wrangling

```python
# Convert JSON objects in 'TOTAL' dataset to Python dicts using json.loads function.
# Then, store the dicts in a list.
# Note that readlines() method is used because .txt file contains multiple, separate JSON objects.
# (pandas's pd.read_json() function only works on single JSON objects.)

path = local_path + '/TOTAL.txt'
with open(path, 'rb') as f:
    raw_txt = f.readlines()

data = []
for line in raw_txt:
    data.append(json.loads(line))
```

```python
# Examine one sample dict from the list of dicts.

data[0]
```

```python
# The dict contains a total of 1,723 entries. According to EIA's website, this includes: 
# "...all the dataset's series (ordered by most recent update data first)
# followed by all of the dataset's categories (ordered by hierarchy from top to bottom)."

len(data)
```

```python
# Brute force way to figure out how many of the JSON objects are related to data series, rather than categories.
# Not ideal, but quick and dirty...

print data[0].keys(), '\n\n'

# Last series dict entry.
print data[1619].keys(), '\n\n'

# First category dict entry.
print data[1620].keys(), '\n\n'
```

```python
data = data[0:1620]
```

```python
for i in range(len(data)):
    name = data[i]['name']
    units = data[i]['units']
    full_name = name + ', ' + units
    data[i]['full_name'] = str(full_name)

print data[0]['full_name']
```

```python
type(data[0]['data'][0][1])
```

```python
# Create dict to store all the dataset's data by the concat of the data category's name and unit of measurement.
# The concat ensures that each dict key is unique, since some different data collections share a category name,
# and only differ in terms of unit of measurement.

full_list = {}
for i in range(len(data)):
    arr = np.array(data[i]['data'])
    series = pd.Series(arr[:,1], index=arr[:,0])
    full_list[data[i]['full_name']] = series
```

```python
# The code in the previous cell converts all the contents of the dataset to unicode, which pandas can't read.
# Go through full_list, check to see if individual data points are numeric, and, if so, convert from unicode to float.

numbers = ('1 2 3 3 4 5 8 9 0').split()
numbers

for x in full_list:
    for i, y in enumerate(full_list[x]):
        if any(number in y for number in numbers):
            full_list[x][i] = float(y)
```

```python
type(full_list['Distillate Fuel Oil Consumed by the Residential Sector, Annual, Thousand Barrels per Day'][0])
```

```python
# Examine the full list of data category names as dict keys.

len(full_list.keys())
```

```python
df = pd.DataFrame(full_list)
df = df.T
df.head()
```

```python
len(df.index)
```

```python
import datetime

time_parsed_cols = []
for col in df.columns:
    if len(col) > 4:
        col = datetime.datetime.strptime(col, '%Y%m').strftime('%Y %m')
    else:
        col = datetime.datetime.strptime(col, '%Y').strftime('%Y')
    time_parsed_cols.append(col)

time_parsed_cols
```

```python
df.columns = time_parsed_cols
```

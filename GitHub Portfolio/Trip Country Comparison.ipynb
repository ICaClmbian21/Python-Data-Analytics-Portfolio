```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Save country codes in a list used to import csv data and, later on, to refer to country-specific DataFrames.

country_codes = ['AUS', 'CHN', 'IND', 'IDN', 'JPN', 'KHM', 'KOR', 'MMR', 'NZL', 'THA', 'USA', 'UAE']
len(country_codes)
```

```python
# Import each country's csv data, clean and organize it into a DataFrame, and save to a master dict for all countries.

dfs = {}
for country in country_codes:
    path1 = '~/Google Drive/Data Analytics Files/Personal Projects/Country Comparison/CSVs/API_'
    path2 = '_DS2_en_csv_v2.csv'
    path3 = path1 + country + path2
    country_df = pd.read_csv(path3, skiprows=4)
    country_df = country_df.drop(['Unnamed: 60', 'Indicator Code', 'Country Code'], axis=1)
    country_df = country_df.set_index('Indicator Name')
    country_df = country_df.T
    dfs[country] = country_df
```

```python
# Get a full list of all the parameters (i.e., columns) in each country's DataFrame (total columsn = 1,420).

for column in dfs['UAE'].columns:
    print column
```

```python
# Select one parameter as an example for a side-by-side comparison across all countries.

total_bilat_aid = pd.Series()
for df in dfs:
    total_bilat_aid[df]= dfs[df]['Net bilateral aid flows from DAC donors, Total (current US$)'].ix['1960':].sum()
```

```python
# Sorted list shows comparison. India has received the most total bilateral aid from various countries since 1960,
# receiving a total of USD ~55 billion.

total_bilat_aid.fillna(0).sort_values(ascending=False)
```

```python
# Simple horizontal bar graph to visualize above comparison.

total_bilat_aid.fillna(0).sort_values().plot(kind='barh')
```

```python
# Generate list of all parameters related to GDP.

for column in dfs['UAE'].columns:
    if 'GDP' in column:
        print column
```

```python
# Create dict to store data on GDP at PPP (Purchasing Power Parity) for all countries.

gdp_ppp = {}
for df in dfs:
    gdp_ppp[df] = dfs[df]['GDP, PPP (current international $)'].ix['2000':'2014'].fillna(0)
```

```python
# Create list of 12 colors to be used in graphical comparison.

colors = 'blue green red cyan magenta yellow black orange purple peru pink plum'
colors = colors.strip().split()
colors
```

```python
for i, country in enumerate(gdp_ppp):
    if gdp_ppp[country].all():
        gdp_ppp[country].plot(figsize=(15,15), color=colors[i], legend=True, label=country)
```

```python
# Show the GDP at PPP for China and the US for the year 2014,
# the year the size of China's economy eclipsed that of the US.
# In 2014, China's GDP at PPP totaled ~18 trillion USD, compared to ~17 trillion USD for the US.

print gdp_ppp['CHN'][gdp_ppp['USA'] < gdp_ppp['CHN']]
print gdp_ppp['USA'][gdp_ppp['USA'] < gdp_ppp['CHN']]
```
